```python
import pandas as pd
import numpy as np
import matplotlib as plt
data=pd.read_csv('F:\jupyter\startup.csv')
```

```python
X=data.iloc[:,:4]
y=data.iloc[:,-1]
X.head()
y.head()
```

```python
cat=pd.DataFrame(X.drop(['R&D Spend','Administration','Marketing Spend'],axis=1))
from sklearn.preprocessing import OneHotEncoder
hot=OneHotEncoder()
hot.fit(cat)
caat=hot.transform(cat)
cat=pd.DataFrame(caat.toarray(),dtype=int)
```

```python

numerical=pd.concat([X,cat],axis=1)
numerical
```

```python
numerical.drop(labels='State',axis=1,inplace=True)
numerical
```

```python
X=numerical
X=X.iloc[:,:5]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.10)
```

```python
from sklearn.preprocessing import StandardScaler
slr=StandardScaler()
slr.fit(x_train)
x_train=slr.fit_transform(x_train)
x_test=slr.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_predict=reg.predict(x_test)
y_predict
```
