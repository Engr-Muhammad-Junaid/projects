```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
#hue is used to determine the color of line on the basis of attributes(events)
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
#adding the style 
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event')
plt.show()
```

```python
#adding a marker
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event',markers=True)
plt.show()
```

```python
import pandas as pd
pokemon=pd.read_csv('F:\jupyter\pokemon.csv')
pokemon
```

```python
#seaborn bar plot
#so the 0 mean not legendarly and 1 mean legendrly the speed shown on y axis
sns.barplot(x='is_legendary',y='speed',data=pokemon,hue='is_legendary')
plt.show()
```

```python
#bar plot to compare the weight of legendarly and non legendarly 
sns.barplot(x='is_legendary',y='weight_kg',data=pokemon)
plt.show()
```

```python
#now in this bar we have generation for each legendary and non legendary pokemon we have total 7Gen
#we can also use pallete for coloring
sns.barplot(x='is_legendary',y='speed',data=pokemon,hue='generation')
plt.show()
```

```python
#searborn histogram/displot
diamonds=pd.read_csv('F:\jupyter\diamonds.csv')
diamonds.head()
```

```python
#distributio plot is a combination of frequency curve and histogram
sns.distplot(diamonds.price)
plt.show()
```

```python
sns.distplot(diamonds.price,hist=False)
plt.show()
```

```python
#we have just the frequency curve
#we can also shown this on vertical by just using vertical=True
sns.distplot(diamonds.price,color='#3CE0FA',bins=10,kde=False,hist_kws=dict(edgecolor="black"))
plt.show()
```
