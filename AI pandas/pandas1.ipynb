# Concatenating Dataframes
To combine two data frame vertical or horizontaly 
lets first try to join two data frame vertical

```python
first=data[(data['pclass']==1)]
_2nd=data[(data['pclass']==2)]
print(first.shape)
print(_2nd.shape)
#These two data frame have same number of columns it's must to mention before combining two dataframes
#verticaly 
```

```python
final=pd.concat([first,_2nd],axis=1)
final.columns
print(final.shape)
```

```python
#combine two dataframes horizontly we can use append method
comb=first.append(_2nd,ignore_index=True)
comb.head()
```

```python
#the other method is concat to combine two data frames is horizontly
df1=final[:200]
print(df1.shape)
df2=final[200:]
print(df2.shape)
```

```python
#to concante data frame horizantoly you have to same number of rows in data frames
final_data=pd.concat([df1,df2],axis=1,ignore_index=True)
```
