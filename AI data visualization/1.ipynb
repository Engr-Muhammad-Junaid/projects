```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x_vals=np.linspace(0,20,20,dtype=int)
y_vals=[math.sqrt(i) for i in x_vals]
x_vals
```

```python
plt.plot(x_vals,y_vals)
plt.rcParams['figure.figsize']=[3,3]
plt.style.use('fivethirtyeight')
plt.grid()
```

```python
fig=plt.figure()
ax=plt.axes()
x=[i**2 for i in range(1,10)]
y=[i**2 for i in range(1,10)]
plt.plot(x_vals,y_vals,'b',label='sqrt')
plt.plot(x,y,'r',label='sqaure')
plt.legend(loc='upper center')
```

```python
plt.style.use('fivethirtyeight')
plt.grid()
plt.rcParams['figure.figsize']=[5,6]
labels=['IT','Marketing','Data Science','Finance']
values=[500,156,300,510] 
explode=(0.05,0.05,0.05,0.05)
plt.pie(values,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True)
plt.show()
```
