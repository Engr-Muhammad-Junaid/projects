```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime as dt
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('F:\jupyter\covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
df.head(4)
```

```python
#here only taking the specfics columns
df=df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
#renaming the columns
df.columns=['date','state','cured','deaths','confirmed']
```

```python
df[df['confirmed']>20000]
```

```python

df[(df['state']=='Maharashtra') & (df['confirmed'] < 80)]
today=df[(df['date']=='2021-07-17') | (df['state']=='Bihar')]
today
```

```python
today=df[df.date=='2021-11-08']
today
```

```python
max_confirmed=today.sort_values(by='confirmed',ascending=False)
maxx=max_confirmed[:5]
```

```python
sns.set(rc={"figure.figsize":(15,10)})
sns.barplot(x='state',y="confirmed",data=maxx,hue="state")
plt.show()
```

```python

today=df[(df['deaths']) & (df['date']=='2021-11-08')]
max_deaths=today.sort_values(by='deaths',ascending=False)
maxxx=max_deaths[:5]
sns.barplot(x='state',y='deaths',data=maxxx,hue='state')
plt.tight_layout()
plt.title('max deaths for top 5 city ')
plt.show()

```
