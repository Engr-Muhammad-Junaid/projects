```python
#Pandas DataFrame dropna() function is used to remove rows and columns 
#with Null/NaN values. By default, this function returns a new DataFrame and the 
#source DataFrame remains unchanged.
#We can create null values using None, pandas.NaT, and numpy.nan variables.
import pandas as pd
import numpy as np
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing',]
}
```

```python
df =pd.DataFrame(people)
df
```

```python
#Drop missing values.
df.dropna(axis=0,how='any',subset=['age','email'])
```

```python
df.dropna(axis='index', how='all', subset=['last', 'email'])
```

```python
df.isna()
```

```python
#fill None or NUll and Nan to zero
df.fillna(0)
```

```python
df.dtypes
```

```python
#here we want to find the mean of the age columns which contain some string values will do here
#type casting 
df['age'] = df['age'].astype(float)
```

```python

df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
df['age'].mean()
```
