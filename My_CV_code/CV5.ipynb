```python
#drawin on the image
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank_img=np.zeros(shape=(512,512,3),dtype=np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
#cv2.rectangle(blank_img,pt1=(230,200),pt2=(450,400),color=(255,0,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,200,0),thickness=10)
plt.imshow(blank_img)

```

```python

cv2.circle(blank_img,center=(200,400),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,pt1=(150,450),pt2=(400,100),color=(0,255,0),thickness=5)
plt.imshow(blank_img)
```

```python
font=cv2.FONT_ITALIC
cv2.putText(blank_img,text='Hello',org=(100,100),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
font=cv2.FONT_HERSHEY_COMPLEX_SMALL
cv2.putText(blank_img,text='Hello',org=(300,450),fontFace=font,fontScale=2,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
#draw your own custom polygin
blank_pic=np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(blank_pic)
```

```python
vertices=np.array([[200,100],[300,200],[100,200]])
vertices              
```

```python
#two steps process.first reshape your vertices 
#2nd pass your point as a list
```

```python
pts=vertices.reshape((-1,1,2))
pts
```

```python
#pass the list of the points array
cv2.polylines(blank_pic,pts=[pts],isClosed=True,color=(0,255,0),thickness=10)
plt.imshow(blank_pic)
```
