```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img =cv2.imread('C://Users//Muhammad junaid//OneDrive//Pictures//mummy_cv.png' ,0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
def img_show(img):
    fig=plt.figure(figsize=(15,15))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    
    
    
```

```python
img_show(img)
```

```python
ret,thur1=cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
```

```python
#here we are going to set each pixel value if the pixel value is greater than the threshold value then it's become white(255).if it's 
#less than the threshold then it will become 0(black) using binary image concept
```

```python
img_show(thur1)
#for each pixel then will be 0 or 255(it will be pure white or pure black)
```
